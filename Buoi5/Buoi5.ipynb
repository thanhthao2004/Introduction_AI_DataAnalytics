```python
import pandas as pd

year = [2010,2011,2012,2013,2014,2015]

salaryHCM = [8.5,9.0,10.5,12.0,11.0,10.0]

salaryLA = [5.5,6.0,6.0,5.5,6.0,5.0]

df = pd.DataFrame(list(zip(year, salaryHCM,salaryLA)),
                  columns=['Year', 'HCM','LA'])
df
```

### 2.1 Nomolization

### Công thức MAX ABSOLUTE
### [-1,1] = x(all)/|max|

```python
max_la = df['LA'].abs().max()
df['macLA'] = df['LA'] / max_la
df
```

```python
max_HCM = df['HCM'].abs().max()
df['macHCM'] = df['HCM'] / max_HCM
df
```

```python
from sklearn.preprocessing import MaxAbsScaler
scaler = MaxAbsScaler()
scaled = scaler.fit_transform(df[['HCM','LA']])
scaled_df = pd.DataFrame(scaled, columns=df[['HCM','LA']].columns)

print(scaled_df)
```

```python
#hãy cho biết số lượng scaler của 4.5 và 5.75
vSc = scaler.transform([[4.5,5.75]])
vSc
```

```python
#Nếu biết trước giá trị sau khi scale là 0.5 và 0.8 thì giá trị gốc tương ứng là:
vR = scaler.inverse_transform([[0.5,0.8]])
vR
```

### Công thức (x(cột)-min)/max-min

```python
#hàm min max scale, giá trị cũ đi qua hàm này sẽ trở về đoạn [0,1]
min_la = df['LA'].min()
max_la = df['LA'].max()
df['mmcLA'] = (df['LA'] - min_la) / (max_la - min_la)

df
```

```python
#min-max scale cho HCM
min_HCM = df['HCM'].min()
max_HCM = df['HCM'].max()
df['mmcHCM'] = (df['HCM'] - min_HCM) / (max_HCM - min_HCM)
df
```

```python
#giả sử ta có lương HCM và LA
vSc = scaler.transform([[6.5,19]])
vSc
#kết quả sai vì bị tràn số
```

```python
vR = scaler.inverse_transform([[0.65,1.9]])
vR
```
